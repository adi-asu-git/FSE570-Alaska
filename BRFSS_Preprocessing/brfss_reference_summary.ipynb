# BRFSS Cleaned Dataset Reference

This notebook summarizes the cleaned BRFSS 2020–2024 dataset used in the project.

```python
main_cols = [
    "_STATE",
    "_LLCPWT",
    "_BMI5",
    "_AGEG5YR",
    "_SEX",
    "_EDUCAG",
    "_INCOMG1",
    "_RACE",
    "BMI",
    "obese",
    "year"
]

df[main_cols].head()
```

```python
age_map = {
    1: "18-24",
    2: "25-29",
    3: "30-34",
    4: "35-39",
    5: "40-44",
    6: "45-49",
    7: "50-54",
    8: "55-59",
    9: "60-64",
    10: "65-69",
    11: "70-74",
    12: "75-79",
    13: "80+"
}

sex_map = {
    1: "Male",
    2: "Female"
}

education_map = {
    1: "Did not graduate high school",
    2: "Graduated high school",
    3: "Attended college or technical school",
    4: "Graduated college or technical school"
}

income_map = {
    1: "<15k",
    2: "15k-25k",
    3: "25k-35k",
    4: "35k-50k",
    5: "50k+"
}

race_map = {
    1: "White",
    2: "Black",
    3: "Other / Multiracial",
    4: "Asian",
    5: "Hispanic"
}
```

```python
df["age_group"] = df["_AGEG5YR"].map(age_map)
df["sex"] = df["_SEX"].map(sex_map)
df["education"] = df["_EDUCAG"].map(education_map)
df["income_group"] = df["_INCOMG1"].map(income_map)
df["race_group"] = df["_RACE"].map(race_map)
```

```python
print("Age groups:")
print(sorted(df["age_group"].dropna().unique()))

print("\nSex groups:")
print(sorted(df["sex"].dropna().unique()))

print("\nEducation groups:")
print(sorted(df["education"].dropna().unique()))

print("\nIncome groups:")
print(sorted(df["income_group"].dropna().unique()))

print("\nRace groups:")
print(sorted(df["race_group"].dropna().unique()))
```

```python
group_cols = ["age_group", "sex", "education", "income_group", "race_group"]

brfss_group_summary = (
    df.groupby(group_cols)
      .agg(
          obesity_rate=("obese", "mean"),
          n=("obese", "size")
      )
      .reset_index()
      .sort_values("n", ascending=False)
)

brfss_group_summary.head(20)
```

## Final Summary of BRFSS Dataset (2020–2024)

The cleaned BRFSS dataset consists of approximately **1.24 million observations** across five years (2020–2024), with a consistent set of demographic and health-related variables used for modeling obesity prevalence. The dataset includes key features such as age group, sex, education level, income group, and race, along with a derived BMI variable and a binary obesity indicator.

Across all years, the dataset shows stable sample sizes (~230k–300k per year) and an overall obesity prevalence of approximately **35.4%**, with a noticeable increase from 2020 (~33%) to later years (~36%). Demographic distributions are well-balanced, with slightly more female respondents (~53%) and a majority of individuals falling into higher income and education categories.

Clear patterns emerge from the exploratory analysis:
- Obesity prevalence increases with age, peaking around 45–54 and declining in older age groups.
- Females show slightly higher obesity prevalence (~36.8%) compared to males (~33.8%).
- Lower income and lower education groups have higher obesity rates, indicating strong socioeconomic gradients.
- Racial disparities are evident, with the highest prevalence observed among Black individuals (~44.9%) and the lowest among Asians (~19.6%).

The dataset was further structured into demographic groups defined by combinations of age, sex, education, income, and race. These group-level summaries (including obesity rates and sample sizes) form the foundation for downstream modeling and will be used in conjunction with ACS demographic distributions for county-level estimation.
