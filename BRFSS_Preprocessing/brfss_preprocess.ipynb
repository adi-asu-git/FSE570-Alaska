```python
def preprocess_brfss(file, year):

    df = pd.read_sas(file, format="xport", encoding="latin1")

    # Fix income variable (2020)
    if year == 2020:
        df = df.rename(columns={"_INCOMG": "_INCOMG1"})

    # Fix race variable (2022+)
    if year >= 2022:
        df = df.rename(columns={"_RACEGR4": "_RACE"})

    cols_needed = [
        "_STATE",
        "_LLCPWT",
        "_BMI5",
        "_AGEG5YR",
        "_SEX",
        "_EDUCAG",
        "_INCOMG1",
        "_RACE"
    ]

    df = df[cols_needed]

    # Convert BMI
    df["BMI"] = df["_BMI5"] / 100

    # Obesity indicator
    df["obese"] = (df["BMI"] >= 30).astype(int)

    # Remove invalid BMI
    df = df[(df["_BMI5"] < 9000) & (df["_BMI5"] > 1200)]

    # Remove missing demographics
    df = df[
        (df["_AGEG5YR"] < 14) &
        (df["_SEX"] < 3) &
        (df["_EDUCAG"] < 5) &
        (df["_INCOMG1"] < 6)
    ]

    df["year"] = year

    return df
```

```python
brfss_2021 = preprocess_brfss("LLCP2021.XPT", 2021)
brfss_2022 = preprocess_brfss("LLCP2022.XPT", 2022)
brfss_2023 = preprocess_brfss("LLCP2023.XPT", 2023)
brfss_2024 = preprocess_brfss("LLCP2024.XPT", 2024)
```

```python
print(brfss_2021.shape)
print(brfss_2022.shape)
print(brfss_2023.shape)
print(brfss_2024.shape)
```

```python
brfss_all = pd.concat([
    brfss_2020,
    brfss_2021,
    brfss_2022,
    brfss_2023,
    brfss_2024
], ignore_index=True)

brfss_all.shape
```
