```python
import pandas as pd

def process_race(file_path, year):
    df = pd.read_csv(file_path, skiprows=1)

    # keep only what we need
    df = df.rename(columns={
        "Geography": "GEO_ID",
        "Geographic Area Name": "NAME"
    })

    def get_col(text):
        matches = [c for c in df.columns if text in c]
        if not matches:
            raise KeyError(f"Missing column containing: {text}")
        return matches[0]

    df["Total"] = pd.to_numeric(df[get_col("Estimate!!Total:")], errors="coerce")
    df["NH_White"] = pd.to_numeric(df[get_col("Not Hispanic or Latino:!!White alone")], errors="coerce")
    df["NH_Black"] = pd.to_numeric(df[get_col("Not Hispanic or Latino:!!Black or African American alone")], errors="coerce")
    df["AIAN"] = pd.to_numeric(df[get_col("Not Hispanic or Latino:!!American Indian and Alaska Native alone")], errors="coerce")
    df["Asian"] = pd.to_numeric(df[get_col("Not Hispanic or Latino:!!Asian alone")], errors="coerce")
    df["NHOPI"] = pd.to_numeric(df[get_col("Not Hispanic or Latino:!!Native Hawaiian and Other Pacific Islander alone")], errors="coerce")
    df["Other"] = pd.to_numeric(df[get_col("Not Hispanic or Latino:!!Some other race alone")], errors="coerce")
    df["Multiracial"] = pd.to_numeric(df[get_col("Not Hispanic or Latino:!!Two or more races")], errors="coerce")
    df["Hispanic"] = pd.to_numeric(df[get_col("Hispanic or Latino")], errors="coerce")

    race_cols = ["NH_White", "NH_Black", "AIAN", "Asian", "NHOPI", "Other", "Multiracial", "Hispanic"]

    df_long = df.melt(
        id_vars=["GEO_ID", "NAME"],
        value_vars=race_cols,
        var_name="race",
        value_name="population"
    )

    df_long["year"] = year
    df_long["prop"] = df_long["population"] / df_long.groupby("GEO_ID")["population"].transform("sum")

    return df_long
```

```python
import pandas as pd

def process_education(file_path, year):
    df = pd.read_csv(file_path, skiprows=1)

    df = df.rename(columns={
        "Geography": "GEO_ID",
        "Geographic Area Name": "NAME"
    })

    def get_col(text):
        matches = [c for c in df.columns if text in c]
        if not matches:
            raise KeyError(f"Missing: {text}")
        return matches[0]

    # Build bins
    df["<HS"] = df[[c for c in df.columns if any(x in c for x in [
        "No schooling", "Nursery", "Kindergarten",
        "1st grade","2nd grade","3rd grade","4th grade",
        "5th grade","6th grade","7th grade","8th grade",
        "9th grade","10th grade","11th grade",
        "12th grade, no diploma"
    ])]].apply(pd.to_numeric, errors="coerce").sum(axis=1)

    df["HS_grad"] = (
        pd.to_numeric(df[get_col("Regular high school diploma")], errors="coerce") +
        pd.to_numeric(df[get_col("GED or alternative credential")], errors="coerce")
    )

    df["Some_college"] = df[[c for c in df.columns if any(x in c for x in [
        "Some college, less than 1 year",
        "Some college, 1 or more years, no degree",
        "Associate"
    ])]].apply(pd.to_numeric, errors="coerce").sum(axis=1)

    df["College_grad"] = df[[c for c in df.columns if any(x in c for x in [
        "Bachelor",
        "Master",
        "Professional school",
        "Doctorate"
    ])]].apply(pd.to_numeric, errors="coerce").sum(axis=1)

    # long format
    df_long = df.melt(
        id_vars=["GEO_ID","NAME"],
        value_vars=["<HS","HS_grad","Some_college","College_grad"],
        var_name="education",
        value_name="population"
    )

    df_long["year"] = year
    df_long["prop"] = df_long["population"] / df_long.groupby("GEO_ID")["population"].transform("sum")

    return df_long
```

```python
import pandas as pd
import re

def process_income(file_path, year):
    df = pd.read_csv(file_path, skiprows=1)

    df = df.rename(columns={
        "Geography": "GEO_ID",
        "Geographic Area Name": "NAME"
    })

    def norm(s):
        s = str(s).lower()
        s = re.sub(r"[^a-z0-9]+", "", s)
        return s

    def find_col(fragment):
        frag = norm(fragment)
        matches = [c for c in df.columns if frag in norm(c)]
        if not matches:
            raise KeyError(f"Missing column containing: {fragment}")
        return matches[0]

    total_col = find_col("estimate total")
    df["Total"] = pd.to_numeric(df[total_col], errors="coerce")

    low_frags = [
        "less than $10,000",
        "$10,000 to $14,999",
        "$15,000 to $19,999",
        "$20,000 to $24,999",
        "$25,000 to $29,999",
        "$30,000 to $34,999"
    ]

    mid_frags = [
        "$35,000 to $39,999",
        "$40,000 to $44,999",
        "$45,000 to $49,999",
        "$50,000 to $59,999",
        "$60,000 to $74,999",
        "$75,000 to $99,999"
    ]

    high_frags = [
        "$100,000 to $124,999",
        "$125,000 to $149,999",
        "$150,000 to $199,999",
        "$200,000 or more"
    ]

    df["Low"] = sum(
        pd.to_numeric(df[find_col(frag)], errors="coerce")
        for frag in low_frags
    )

    df["Mid"] = sum(
        pd.to_numeric(df[find_col(frag)], errors="coerce")
        for frag in mid_frags
    )

    df["High"] = sum(
        pd.to_numeric(df[find_col(frag)], errors="coerce")
        for frag in high_frags
    )

    df_long = df.melt(
        id_vars=["GEO_ID", "NAME"],
        value_vars=["Low", "Mid", "High"],
        var_name="income",
        value_name="population"
    )

    df_long["year"] = year
    df_long["prop"] = df_long["population"] / df_long.groupby("GEO_ID")["population"].transform("sum")

    return df_long
```
