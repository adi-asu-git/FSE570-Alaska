```python
import pandas as pd
import numpy as np

files = {
    2020: "LLCP2020.XPT",
    2021: "LLCP2021.XPT",
    2022: "LLCP2022.XPT",
    2023: "LLCP2023.XPT",
    2024: "LLCP2024.XPT"
}

for year, f in files.items():
    df_temp = pd.read_sas(f, format="xport", encoding="latin1")
    income_cols = [c for c in df_temp.columns if "INCOM" in c]
    race_cols = [c for c in df_temp.columns if "RACE" in c]
    print(f"\n{year}:")
    print(f"  Income: {income_cols}")
    print(f"  Race: {race_cols}")
```

```python
print("=" * 60)
print("INCOME VARIABLE ANALYSIS")
print("=" * 60)

for year, f in files.items():
    df_temp = pd.read_sas(f, format="xport", encoding="latin1")
    
    if year == 2020:
        col = "_INCOMG"
    else:
        col = "_INCOMG1"
    
    print(f"\n{year} — {col}:")
    print(df_temp[col].value_counts().sort_index())
```

```python
print("=" * 60)
print("RACE VARIABLE ANALYSIS")
print("=" * 60)

for year, f in files.items():
    df_temp = pd.read_sas(f, format="xport", encoding="latin1")
    
    if year == 2022:
        race_col = "_RACEPR1"
        raw_col = "_RACE1"
    else:
        race_col = "_RACEPRV"
        raw_col = "_RACE"
    
    print(f"\n{year} — raw: {raw_col}, calculated: {race_col}")
    print(f"  {raw_col}:", df_temp[raw_col].value_counts().sort_index().to_dict())
    print(f"  {race_col}:", df_temp[race_col].value_counts().sort_index().to_dict())
```

```python
print("2020 _INCOMG bins (5 categories):")
print("  1: <15k")
print("  2: 15k-25k") 
print("  3: 25k-35k")
print("  4: 35k-50k")
print("  5: 50k+ (overloaded — captures all income above 50k)")

print("\n2021-2024 _INCOMG1 bins (7 categories):")
print("  1: <15k")
print("  2: 15k-25k")
print("  3: 25k-35k")
print("  4: 35k-50k")
print("  5: 50k-100k")
print("  6: 100k-200k")
print("  7: 200k+")

print("\nRespondents in high income brackets:")
print(f"  2020 code 5 (50k+):     171,265")
print(f"  2021 codes 5+6+7 total: {107246+67411+18952:,}")
print(f"  2022 codes 5+6+7 total: {107584+72883+23478:,}")
print(f"  2023 codes 5+6+7 total: {107027+76637+26770:,}")
print(f"  2024 codes 5+6+7 total: {112409+82453+29694:,}")
```

```python
print("=" * 60)
print("INCOME: COST/BENEFIT ANALYSIS")
print("=" * 60)

# How many rows are in 2020 income data
df_2020 = pd.read_sas("LLCP2020.XPT", format="xport", encoding="latin1")
total_2020 = len(df_2020)
valid_2020_income = len(df_2020[df_2020["_INCOMG"] < 9])
print(f"\n2020 total rows: {total_2020:,}")
print(f"2020 rows with valid income: {valid_2020_income:,}")
print(f"2020 rows lost if we drop income: {valid_2020_income:,} lose income label")

# Option A: Drop 2020 income entirely
print("\n--- Option A: Drop 2020 income (use NaN) ---")
print(f"  Rows kept: all {total_2020:,} from 2020 but income = NaN")
print(f"  Income analysis uses 2021-2024 only")
total_2021_2024 = 438693 + 445132 + 433323 + 457670
print(f"  Total rows for income analysis: {total_2021_2024:,}")

# Option B: Collapse all years to 5 bins
print("\n--- Option B: Collapse to 5 bins (50k+) for all years ---")
print(f"  All 5 years usable for income analysis")
print(f"  Total rows: {total_2020 + total_2021_2024:,}")
print(f"  BUT: high income detail lost for 2021-2024")

# Who is in the high income brackets (codes 5,6,7 in 2021-2024)
print("\n--- Who is in codes 5,6,7 (50k+) in 2021-2024? ---")
for year, f in {2021: "LLCP2021.XPT", 2022: "LLCP2022.XPT", 
                2023: "LLCP2023.XPT", 2024: "LLCP2024.XPT"}.items():
    df_temp = pd.read_sas(f, format="xport", encoding="latin1")
    high_income = df_temp[df_temp["_INCOMG1"].isin([5, 6, 7])]
    total = len(df_temp[df_temp["_INCOMG1"] < 9])
    pct = len(high_income) / total * 100
    print(f"  {year}: {len(high_income):,} of {total:,} valid rows ({pct:.1f}%) earn 50k+")
```

```python
print("=" * 60)
print("INCOME: DEMOGRAPHIC IMPACT OF COLLAPSING HIGH INCOME BINS")
print("=" * 60)

# Use 2021-2024 combined to show who is in codes 5, 6, 7
dfs = []
for year, f in {2021: "LLCP2021.XPT", 2022: "LLCP2022.XPT",
                2023: "LLCP2023.XPT", 2024: "LLCP2024.XPT"}.items():
    df_temp = pd.read_sas(f, format="xport", encoding="latin1")
    if year == 2022:
        df_temp = df_temp.rename(columns={"_RACEPR1": "_RACEPRV"})
    df_temp["year"] = year
    dfs.append(df_temp[["_INCOMG1", "_EDUCAG", "_AGEG5YR", "_RACEPRV", "year"]])

combined = pd.concat(dfs, ignore_index=True)
combined = combined[combined["_INCOMG1"] < 9]

# High income = codes 5, 6, 7
combined["high_income"] = combined["_INCOMG1"].isin([5, 6, 7])

print("\nBy Education (% in high income brackets):")
educ_map = {1: "No HS", 2: "HS grad", 3: "Some college", 4: "College grad"}
combined["educ"] = combined["_EDUCAG"].map(educ_map)
print(combined.groupby("educ")["high_income"].mean().sort_values(ascending=False).round(3))

print("\nBy Race (% in high income brackets):")
race_map2 = {1: "NH-White", 2: "NH-Black", 3: "AIAN", 
             4: "Asian", 5: "NHOPI", 6: "Other/Multiracial", 7: "Hispanic"}
combined["race"] = combined["_RACEPRV"].map(race_map2)
print(combined.groupby("race")["high_income"].mean().sort_values(ascending=False).round(3))

print("\nBy Age (% in high income brackets):")
age_map2 = {1:"18-24",2:"25-29",3:"30-34",4:"35-39",5:"40-44",
            6:"45-49",7:"50-54",8:"55-59",9:"60-64",10:"65-69",
            11:"70-74",12:"75-79",13:"80+"}
combined["age"] = combined["_AGEG5YR"].map(age_map2)
print(combined.groupby("age")["high_income"].mean().sort_index().round(3))
```

```python
print("=" * 60)
print("RACE: _RACE vs _RACEPRV IMPUTATION ANALYSIS")
print("=" * 60)

race_labels = {
    1: "NH-White", 2: "NH-Black", 3: "AIAN",
    4: "Asian", 5: "NHOPI", 6: "Other",
    7: "Multiracial", 8: "Hispanic"
}

for year, f in {2020: "LLCP2020.XPT", 2021: "LLCP2021.XPT",
                2023: "LLCP2023.XPT", 2024: "LLCP2024.XPT"}.items():
    df_temp = pd.read_sas(f, format="xport", encoding="latin1")
    
    # rows where _RACE is missing but _RACEPRV has a value
    missing_race = df_temp[df_temp["_RACE"] == 9]
    total_missing = len(missing_race)
    total_rows = len(df_temp)
    
    print(f"\n{year}:")
    print(f"  Total rows: {total_rows:,}")
    print(f"  _RACE missing (code 9): {total_missing:,} ({total_missing/total_rows*100:.1f}%)")
    print(f"  How _RACEPRV filled those {total_missing:,} missing rows:")
    imputed = missing_race["_RACEPRV"].value_counts().sort_index()
    for code, count in imputed.items():
        label = race_labels.get(int(code), "Unknown")
        print(f"    code {int(code)} ({label}): {count:,} ({count/total_missing*100:.1f}%)")
```
