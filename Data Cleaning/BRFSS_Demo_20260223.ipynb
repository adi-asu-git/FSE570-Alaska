```python
# 2023 Alabama data
al_bmi = df[(df["Locationabbr"] == "AL") & (df["Year"] == 2023)].copy()

al_bmi["Conf_int"] = al_bmi["Confidence_limit_High"]- al_bmi["Data_value"]
al_bmi = al_bmi[["Year",'Locationabbr','Topic','Response','Break_Out','Break_Out_Category','Sample_Size','Data_value','Conf_int']].copy()

al_bmi
```

```python
## 25 was largest sample size that left response as NAN
al_bmi[al_bmi.isna().any(axis=1)]
```

```python
# Split data by Obese (BMI > 30) and Not Obese (Bmi < 30)
bmi_under_30 = [
    "Underweight (BMI 12.0-18.4)",
    "Normal Weight (BMI 18.5-24.9)",
    "Overweight (BMI 25.0-29.9)"
]

al_bmi["Response_Mod"] = "BMI > 30"

al_bmi.loc[al_bmi["Response"].isin(bmi_under_30),"Response_Mod"] = "BMI < 30"
al_bmi.head(20)
```

```python
# Calculate weighted mean of Obese and Not Obese by breakout in each breakout category
al_bmi["Data_value"] = pd.to_numeric(al_bmi["Data_value"])
al_bmi["Sample_Size"] = (
    al_bmi["Sample_Size"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .replace("nan", np.nan)
    .astype(float)
)

# Start with original values
al_bmi["Data_value_Mod"] = al_bmi["Data_value"]

# Mask for BMI < 30
m = al_bmi["Response_Mod"].eq("BMI < 30")

group_cols = ["Break_Out", "Break_Out_Category"]

# Compute weighted means excluding NaNs
wmean = (
    al_bmi.loc[m]
    .groupby(group_cols)[["Data_value", "Sample_Size"]]
    .apply(
        lambda g: np.average(
            g["Data_value"].dropna(),
            weights=g.loc[g["Data_value"].notna(), "Sample_Size"]
        )
        if g["Data_value"].notna().any()
        else np.nan
    )
)

# Assign back to BMI < 30 rows
al_bmi.loc[m, "Data_value_Mod"] = (
    al_bmi.loc[m]
    .set_index(group_cols)
    .index.map(wmean)
    .to_numpy()
)
al_bmi.head(20)
```

```python
# Drop Response Column

bmi_under_30 = [
    "Underweight (BMI 12.0-18.4)",
    "Normal Weight (BMI 18.5-24.9)"
]

al_bmi = al_bmi[~al_bmi["Response"].isin(bmi_under_30)].copy()

al_bmi = al_bmi.drop(columns=["Response"])
al_bmi
```
